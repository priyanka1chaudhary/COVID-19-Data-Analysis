```python
# covid_india_analysis.py

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load dataset
df = pd.read_csv("/content/covid_19_india.csv")

df.columns = df.columns.str.strip()  # clean spaces
print(df.columns)

df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%y')

# Clean column names (optional)
df.columns = df.columns.str.strip().str.replace(' ', '_')

# Preview data
print(df.head())

# ----------------- TIME TREND ANALYSIS ----------------- #

# Aggregate data by Date
daily_totals = df.groupby('Date')[['Confirmed', 'Cured', 'Deaths']].sum().reset_index()

# Plot daily trends
plt.figure(figsize=(12,6))
sns.lineplot(data=daily_totals, x='Date', y='Confirmed', label='Confirmed')
sns.lineplot(data=daily_totals, x='Date', y='Cured', label='Cured')
sns.lineplot(data=daily_totals, x='Date', y='Deaths', label='Deaths')
plt.title("India COVID-19 Daily Trends")
plt.xlabel("Date")
plt.ylabel("Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ----------------- HOTSPOTS BY STATE ----------------- #

# Latest date data
latest_date = df['Date'].max()
latest_df = df[df['Date'] == latest_date]

# Top 10 states by Confirmed cases
top_states = latest_df.sort_values(by='Confirmed', ascending=False).head(10)

# Bar plot
plt.figure(figsize=(10,5))
sns.barplot(data=top_states, y='State/UnionTerritory', x='Confirmed', palette='Reds_r')
plt.title(f"Top 10 COVID-19 Hit States as of {latest_date.date()}")
plt.xlabel("Confirmed Cases")
plt.ylabel("State")
plt.tight_layout()
plt.show()

# ----------------- INTERACTIVE MAP (Optional) ----------------- #

# If you have lat/lon for each state, you can create a choropleth using Plotly
# For now, simple state-wise bar chart with Plotly
fig = px.bar(top_states,
             x='State/UnionTerritory',
             y='Confirmed',
             color='Confirmed',
             title='Top 10 States by Confirmed Cases',
             labels={'Confirmed': 'Confirmed Cases'},
             text='Confirmed')
fig.update_traces(texttemplate='%{text}', textposition='outside')
fig.show()

# ----------------- CORRELATION (OPTIONAL) ----------------- #

# If you have a separate file: state_demographics.csv
# Columns: State, Population, Density, Literacy, GDP_per_capita, etc.
# demographic_df = pd.read_csv("state_demographics.csv")

# Merge and perform correlation
# merged_df = latest_df.merge(demographic_df, left_on='State/UnionTerritory', right_on='State')

# corr = merged_df[['Confirmed', 'Cured', 'Deaths', 'Population', 'GDP_per_capita']].corr()
# sns.heatmap(corr, annot=True, cmap='coolwarm')
# plt.title("Correlation between COVID Metrics and State Demographics")
# plt.show()
```
